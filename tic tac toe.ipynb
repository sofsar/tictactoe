```python
board = {'1': ' ', '2': ' ', '3': ' ',
         '4': ' ', '5': ' ', '6': ' ',
         '7': ' ', '8': ' ', '9': ' '}

def printboard():
    
    print(board['1'] + '|' + board['2'] + '|' + board['3'])
    print('-+-+-')
    print(board['4'] + '|' + board['5'] + '|' + board['6'])
    print('-+-+-')
    print(board['7'] + '|' + board['8'] + '|' + board['9'])
    
printboard()
```

```python
def reset():
    for key in board:
        board[key]= ' '
```

```python
def again():
    playagain = input('Do you want to play again? (y/n) ')
    
    if playagain.lower() == 'y':
        reset()
        game()
    elif playagain.lower() == 'n':
        print('See you next time!')
```

```python
def game():
    
    turn = 'X'
    
    while True:
        printboard()
        print("It's your turn " + turn + ".")
        
        spot = input('Select a spot:')
        
        if int(spot) > 9 or int(spot) < 1:
                print('Please type a valid number.')
                spot = input('Select a spot:')
            
        if board[spot] == ' ':
            board[spot] = turn

        else:
            print('This spot is already taken.\nSelect a spot:')
            continue

        if board['1'] == board['2'] == board['3'] != ' ':
            printboard()
            print("Congratulations, " + turn + "! You won.")
            again()
            break
        elif board['4'] == board['5'] == board['6'] != ' ':
            printboard()
            print("Congratulations, " + turn + "! You won.")
            again()
            break
        elif board['7'] == board['8'] == board['9'] != ' ':
            printboard()
            print("Congratulations, " + turn + "! You won.")
            again()
            break
        elif board['1'] == board['4'] == board['7'] != ' ':
            printboard()
            print("Congratulations, " + turn + "! You won.")
            again()
            break
        elif board['2'] == board['5'] == board['8'] != ' ':
            printboard()
            print("Congratulations, " + turn + "! You won.")
            again()
            break
        elif board['3'] == board['6'] == board['9'] != ' ':
            printboard()
            print("Congratulations, " + turn + "! You won.")
            again()
            break
        elif board['1'] == board['5'] == board['9'] != ' ':
            printboard()
            print("Congratulations, " + turn + "! You won.")
            again()
            break
        elif board['3'] == board['5'] == board['7'] != ' ':
            printboard()
            print("Congratulations, " + turn + "! You won.")
            again()
            break
        elif board['1'] != ' ' and board['2'] != ' ' and board['3'] != ' ' and board['4'] != ' ' and board['5'] != ' ' and board['6'] != ' ' and board['7'] != ' ' and board['8'] != ' ' and board['9'] != ' ':
            print("It's a tie!")
            again()
            break
            
        if turn == 'X':
            turn = 'O'
            
        else:
            turn = 'X'           
    
```

```python
def tutorial():
    print("Welcome to Tic-Tac-Toe!\nThe spots on the board are the following:")
    print('1' + '|' + '2' + '|' + '3')
    print('-+-+-')
    print('4' + '|' + '5' + '|' + '6')
    print('-+-+-')
    print('7' + '|' + '8' + '|' + '9')
    
    start = input("Type play to start! ")
    
    if start.lower() == 'play':
        game()
```
